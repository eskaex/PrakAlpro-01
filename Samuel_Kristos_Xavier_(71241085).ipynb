```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('Values');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
Harga0 = 650000
Aset_emas0 = 25
Harga1 = 685000
Jumlah_harga_beli1 = Aset_emas0 * Harga0
Jumlah_harga_jual1 = Aset_emas0 * Harga1
Profit_rupiah1 = Jumlah_harga_jual1 - Jumlah_harga_beli1
Persentase_profit1 = (Profit_rupiah1/Jumlah_harga_beli1) * 100
print("Keuntungan yang diperoleh Gerard adalah Rp", Profit_rupiah1)
print("Gerard untung sebesar", f'{Persentase_profit1:.2f}', "%")
beli_emas1 = 15
Harga2 = 715000
Jumlah_harga_beli2 = beli_emas1 * Harga1
Aset_emas1 = Aset_emas0 + beli_emas1
Profit_rupiah2 = (Aset_emas1 * Harga2) - (Jumlah_harga_beli1 + Jumlah_harga_beli2)
Persentase_profit2 = (Profit_rupiah2/(Jumlah_harga_beli1 + Jumlah_harga_beli2)) * 100
print("Saat harga emas naik lagi, keuntungan yang diperoleh Gerard adalah Rp", Profit_rupiah2)
print("Gerard untung sebesar", f'{Persentase_profit2:.2f}', "%")
```

```python
modal = 200000000
bunga = 0.1
target = 400000000
butuh_waktu = 0

while modal < target:
  butuh_waktu += 1
  kalkulasi = modal * (1 + bunga)
  hasil_pertahun = kalkulasi - modal
  modal += hasil_pertahun

print("Erika butuh waktu", butuh_waktu, "tahun untuk mendapatkan uang Rp", round(kalkulasi))
```
